```python
def view_values(df,column):
  """Visualiza de forma rapida los valores unicos para la variable"""
  valores = df[column].value_counts(dropna=False)
  percent= valores[0]/df.shape[0] * 100
  print(df[column].value_counts(dropna=False))
  if percent >=80:
    print("!!!!---- ALERTA, concentracion de mas del 80%")
    return column
  return None
```

Revisemos rapidamente los valores unicos para las variables preseleccionadas y comencemos a descartar

```python
mas_80 = list()
for i in preselect_vars:
  print("==============NOMBRE COLUMNA: ", i, "====================================\n")
  data = view_values(df_train_main,i)
  if data is not None:
    mas_80.append(data)
```

Primero necesitamos convertir a numero

```python
def replace_binary(df,column, opchange):
  print(df[column].value_counts(dropna=False))

  new_values = df[column].eq(opchange).mul(1).astype('int')
  return new_values
```

```python
for i in var_servicios:
  df_test_main[i] = replace_binary(df_test_main,i, 'Si')
  df_train_main[i] = replace_binary(df_train_main,i, 'Si')

print(df_test_main[var_servicios].info())
print(df_train_main[var_servicios].info())
```

Imputamos Débil para valores diferentes

```python
def transform_categorics(df, column,values_list,replacer):
  """
  Trata valores categoricos segun lista de deseados
  """
  new_values = np.where(df[column].isin(values_list), df[column], replacer)
  return new_values
```

```python
values_dem_int = ['Media','Fuerte','Débil','Nula']
df_train_main['demanda_interes'] = transform_categorics(df_train_main, 'demanda_interes',values_dem_int,'Débil')
```

```python
values_dem_int = ['Media','Fuerte','Débil','Nula']
df_test_main['demanda_interes'] = transform_categorics(df_test_main, 'demanda_interes',values_dem_int,'Débil')
```

### area
Se realiza tratamiento de todas las variables que tengan 'area'

```python
#todas las variables que tienen 'area'
area_cols = [col for col in df_train_main.columns if ('area' in col)]
area_cols
```

```python
df_train_main['area_construccion']
```

```python
df_train_main['area_valorada']
```

```python
def tratar_texto_to_num(df,i):
  """
    Quita caracteres indeseados en texto para luego convertirse a numero
  """
  new_values = df[i].str.replace(',','.')#solo se reemplazan las , se asumen puntos como decimales segun revision
  return new_values
```

```python
for i in area_cols:
  df_train_main[i] = tratar_texto_to_num(df_train_main,i)
  df_train_main[i] = pd.to_numeric(df_train_main[i], errors='coerce')
  df_train_main[i].fillna(df_train_main[i].median(),inplace=True)
  df_train_main[i]= df_train_main[i].astype('float')
  #print(df_train_main[i].value_counts(dropna=False))

  df_test_main[i] = tratar_texto_to_num(df_test_main,i)
  df_test_main[i] = pd.to_numeric(df_test_main[i], errors='coerce')
  df_test_main[i].fillna(df_test_main[i].median(),inplace=True)
  df_test_main[i]= df_test_main[i].astype('float')
  #print(df_test_main[i].value_counts(dropna=False))
```
