```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", None)
from xgboost import XGBRegressor

from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import explained_variance_score
```

```python
#carga de los dos dataset previamente tratados (train y test)
df_train = pd.read_excel('/content/drive/MyDrive/Personal/test_bancolombia/Prueba Modelo_data/train.xlsx', index_col=0)
df_test = pd.read_excel('/content/drive/MyDrive/Personal/test_bancolombia/Prueba Modelo_data/test.xlsx', index_col=0)
```

```python
#carga dataset resultados
df_base_evaluada = pd.read_csv("/content/drive/MyDrive/Personal/test_bancolombia/Prueba Modelo_data/base_evaluada.csv")
```

```python
train_not_0s = pd.read_excel("/content/drive/MyDrive/Personal/test_bancolombia/Prueba Modelo_data/train_no_zeros_only.xlsx",index_col=0)
```

```python
train_not_0s
```

```python
train_not0s_ni_out = pd.read_excel("/content/drive/MyDrive/Personal/test_bancolombia/Prueba Modelo_data/train_no_zeros_no_out.xlsx",index_col=0)
```

```python
train_not0s_ni_out
```

## XGBOOST 1

```python
all_vals = train_not0s_ni_out.columns.to_list()
```

```python
all_vals.remove('valor_total_avaluo')
all_vals.remove('id')
all_vals.remove('municipio_inmueble')
all_vals.remove('tipo_inmueble')
all_vals.remove('antejardin')
all_vals.remove('unidades')
all_vals.remove('nivel_servicios')
```

```python
x_train = train_not0s_ni_out[all_vals]
```

```python
x_train.info()
```

```python
y_train = train_not0s_ni_out[['valor_total_avaluo']]
```

```python
y_train
```

```python
XGBR = XGBRegressor(max_depth=80,learning_rate=0.01,n_estimators=1000, random_state=55)
```

```python
XGBR.fit(x_train,y_train, verbose=True)
```

```python
def calculate_r2(l, pred_values):
    print('r2_score: ', r2_score(l, pred_values))
def calc_moe(l, pred_values):
    print('mean_square_error: ', mean_squared_error(l, pred_values, squared = False))
    print('mean_absolute_error: ', mean_absolute_error(l, pred_values))
    print('mean_absolute_percentage_error: ', mean_absolute_percentage_error(l, pred_values))
    print('explained_variance_score: ', explained_variance_score(l, pred_values))
```

```python
pred_values = XGBR.predict(x_train)
```

```python
calculate_r2(y_train, pred_values)
calc_moe(y_train, pred_values)
```

Un MAPE de 41%, no tenemos 0's en los datos de entrenamiento. Por tal hacer una iteracion incluyendo los 0's solo podria aumentar el MAPE.

El siguiente modelo incluye los outlieres pero no los 0's

### Dataset original (0's y outlieres)

```python
x_train_original = df_train[x_train.columns.to_list()]
```

```python
pred_values_orig = XGBR.predict(x_train_original)
```

```python
y_train_original = df_train['valor_total_avaluo']
```

```python
calculate_r2(y_train_original, pred_values_orig)
calc_moe(y_train_original, pred_values_orig)
```
