## Model 1: Multiple Linear Regression

```python
#Utilizamos todas las variables de area y algunas que demostraron mayor correlación
set_x = [
 'area_privada',
 'area_garaje',
 'area_terreno',
 'area_construccion',
 'area_deposito',
 'area_otros',
 'area_libre',
 'estrato',
 'alamedas',
 'cuarto_servicio',
 'zona_verde_privada',
 'estar_habitacion',
 'acueducto_en_el_sector',
 'bano_social',
 'nivel_equipamiento_comercial'
]

y = 'valor_total_avaluo'
```

```python
#concatenate values
vars_x = ' + '
vars_x = vars_x.join(set_x)
print(vars_x)
```

```python
formula_str = y + ' ~ ' + vars_x
print(formula_str)
```

```python
model = sm.OLS.from_formula(formula_str, data=df_train)

result = model.fit()
result.summary()
```

Los valores de P>|indican relaciones muy bajas de las variables con el objetivo y el valor de r-cuadrado es muy bajo.

El modelo multiple linear regression propuesto resulta poco efectivo.


```python
df_test['valor_total_avaluo_pred'] = result.predict(df_test[set_x])
```

```python
df_tojoin= df_test[['valor_total_avaluo_pred','id']]
```

```python
response = df_base_evaluada.merge(df_tojoin, on='id')
```
