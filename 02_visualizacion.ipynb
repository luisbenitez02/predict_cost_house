## Distribución variables de area

```python
#veremos la distribucion de un set numerico de variables (no categoricos ni ordinales)
subnumeric_set = [
  'area_privada',
 'area_garaje',
 'area_terreno',
 'area_construccion',
 'area_deposito',
 'area_otros',
 'area_libre',
 'area_valorada'             
]
```

```python
#se utiliza asi debido a errores con seaborn en el entorno de google colab
def graficar_distribution(df,x_column,y_colum):
  plt.figure(dpi = 100)
  plt.scatter(df[x_column],df[y_colum], marker='*');
  plt.title(x_column)
  plt.show()
```

```python
for i in subnumeric_set:
  graficar_distribution(df_train,i,'valor_total_avaluo')
```

```python
plt.figure(dpi = 100)
plt.scatter(df_train['area_valorada'],df_train['valor_total_avaluo'], marker='*');
plt.axis([0, 4000, 0, 30000000000])
plt.title(i)
plt.show()
```
